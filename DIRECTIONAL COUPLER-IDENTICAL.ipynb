# DIRECTIONAL COUPLER

```python
import numpy as np
import sympy as sp
import math
from sympy import Symbol
import matplotlib.pyplot as plt
x=Symbol('x')
#l=float(input("Enter wavelength of incident wave:"))
l=1.5
k0=2*3.14/l
def K(n):                  # wave number
    y=k0*(n**2-x**2)**0.5
    return y
print(K(1))
#layers=int(input("Enter no. of layers in waveguide:"))
#waveguides=int(input("Enter no. of parllel waveguides:"))
layers=3
waveguides=1
ref=[1.49,1.5,1.49]
width=[10.15,1,10.1]
a=[]
b=[]
Z=[]
#for i in range(layers):
    #ref=ref+[float(input("Refractive index:n"+str(i)+"="))]
    #width=width+[float(input("Width for n"+str(i)+"="))]
for j in range(layers-1):
    matrix1=sp.Matrix([[1,0],[0,1]])
    
    for i in range(j,layers-1):
        n1=ref[i]
        n2=ref[i+1]
        d=width[i]

        S=(K(n2)/K(n1))
        L1=(sp.exp(-1j*K(n1)*d))
        L2=sp.exp(1j*K(n1)*d)
        ele11=(0.5*(1+S)*L1)
        ele12=(0.5*(1-S)*L1)
        ele21=(0.5*(1-S)*L2)
        ele22=(0.5*(1+S)*L2)
        matrix2=sp.Matrix([[ele11,ele12],[ele21,ele22]])
        matrix=matrix1*(matrix2)
        matrix1=(matrix)
    a0=matrix[0]
    b0=matrix[2]
    equ=matrix[0]
    a=a+[a0]
    b=b+[b0]
    if j==0:
          equa=(equ)
def fun(y):
    return ((equa.subs(x,y)))
y12=np.arange(ref[0]+(ref[1]-ref[0])/100,ref[1],(ref[1]-ref[0])/100)
z1=[]

for i in y12:
    z1=z1+[sp.re(sp.N(fun(i)))]
for j in range(0,len(y12)-1):
    if (z1[j]*z1[j+1]<0):
        y1=y12[j]
        y2=y12[j+1]

        y3=(y1+y2)/2
        E=(sp.re(sp.N(fun(y3))))
        D=(sp.re(sp.N(fun(y1))))
        while abs(E)>0.001:
            if ((D*E)<=0):
                y2=y3
            else:
                 y1=y3
            y3=(y1+y2)/2
            E=(sp.re(sp.N(fun(y3))))
    
        Z=Z+[y3]   
print(Z)
A=[]
B=[]
for k in range(len(Z)):
    A1=[]
    B1=[]
    for i in range(layers-1):
   
         sp.sigmoid=(a[i])
         VA=sp.N(sp.sigmoid.subs(x,Z[k]))
         sp.sigmoid=(b[i])
         VB=sp.N(sp.sigmoid.subs(x,Z[k]))
         
         A1=A1+[VA]
         B1=B1+[VB]
        
    A1=A1+[1]
    B1=B1+[0]
    A=A+[A1]
    B=B+[B1]
for k in range(len(Z)):
    d=0
    z=[]
    for j in range(layers):
        y=np.arange(d,d+width[j]+0.1,0.1)
        for i in  y:
            sp.sigmoid=(A[k][j]*sp.exp(1j*K(ref[j])*(i-d))+B[k][j]*sp.exp(-1j*K(ref[j])*(i-d)))
            z=z+[sp.re(sp.N(sp.sigmoid.subs(x,Z[k])))]
#         plt.subplot(3,1,k+1)
#         plt.plot(y,z)
#         plt.xlabel('X')
        d = d + width[j]
        #plt.grid()
Y=z
len(Y)
```
